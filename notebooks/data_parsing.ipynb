```python
import numpy as np
import pandas
```

```python
# load data
data_csv = "../../data/warfarin.csv"
data = pandas.read_csv(data_csv, delimiter=',')

cols = data.columns
cols = cols.map(lambda x: x.replace(' ', '_') if isinstance(x, (str, unicode)) else x)
data.columns = cols

print(data.shape)
print(list(data))
```

```python
# filter out rows ina . dataframe with column year values NA/NAN
data_filtered = data[data.Therapeutic_Dose_of_Warfarin.notnull()]

print(data_filtered.shape)
```

```python
# bucket the dosage into 0 for under 3mg/day, 1 for 3-7 mg/day, and 2 for over 7 mg/day
# NOTE the csv has dosage of mg/week

data_filtered['Therapeutic_Dose_of_Warfarin_binned'] = 0
for index, row in data_filtered.iterrows():
    dose_per_day = row["Therapeutic_Dose_of_Warfarin"] / 7.0
    
    if dose_per_day < 3.0:
        data_filtered.at[index, "Therapeutic_Dose_of_Warfarin_binned"] = 0
    elif dose_per_day <= 7.0:
        data_filtered.at[index, "Therapeutic_Dose_of_Warfarin_binned"] = 1
    else:
        data_filtered.at[index, "Therapeutic_Dose_of_Warfarin_binned"] = 2
        
print(data_filtered.shape)

for index, row in data_filtered.iterrows():
    print(row["Therapeutic_Dose_of_Warfarin"])
    print(row["Therapeutic_Dose_of_Warfarin_binned"])
```

```python
# bucket the ages by decade

print(data_filtered.shape[0])
decades = []

for index, row in data_filtered.iterrows():
    age = row["Age"]
        
    if pandas.isnull(age):
        continue
    else:
        decade = float(age[0])
        data_filtered.at[index, "Age"] = decade
        decades.append(decade)
        
print(data_filtered.shape)

for index, row in data_filtered.iterrows():
    print(row["Age"])
```

```python
# now replace the nans with median decade
print(len(decades))
decades.sort()
median = decades[(len(decades)-1)/2]
print(median)

for index, row in data_filtered.iterrows():
    age = row["Age"]
    if pandas.isnull(age):
        data_filtered.at[index, "Age"] = median
        
print(data_filtered.shape)
```

```python
# bucket for VKORC1 G/A, cyp2c9, and race

df = pandas.get_dummies(data_filtered, ["VKORC1"], columns=["VKORC1_genotype:_-1639_G>A_(3673);_chr16:31015190;_rs9923231;_C/T"], dummy_na=True)
df = pandas.get_dummies(df, ["Cyp2C9"], columns=["Cyp2C9_genotypes"], dummy_na=True)
df = pandas.get_dummies(df, ["Race"], columns=["Race"], dummy_na=True)
```

```python
# now we want to do the medications

df['amiodarone_status'] = 0.0
df['enzyme_inducer_status'] = 0.0

for index, row in data_filtered.iterrows():
    amiodarone_status = row["Amiodarone_(Cordarone)"]
    if not pandas.isnull(amiodarone_status) and float(amiodarone_status) == 1:
        data_filtered.at[index, "amiodarone_status"] = 1
    else:
        data_filtered.at[index, "amiodarone_status"] = 0
        
    carbamazepine_status = row["Carbamazepine_(Tegretol)"]
    phenytoin_status = row["Phenytoin_(Dilantin)"]
    rifampin_status = row["Rifampin_or_Rifampicin"]
    
    if not pandas.isnull(carbamazepine_status) and float(carbamazepine_status) == 1:
        data_filtered.at[index, "enzyme_inducer_status"] = 1
    elif not pandas.isnull(phenytoin_status) and float(phenytoin_status) == 1:
        data_filtered.at[index, "enzyme_inducer_status"] = 1
    elif not pandas.isnull(rifampin_status) and float(rifampin_status) == 1:
        data_filtered.at[index, "enzyme_inducer_status"] = 1
    else:
        data_filtered.at[index, "enzyme_inducer_status"] = 0

print(list(df))
            
```

```python
cols = df.columns
cols = cols.map(lambda x: x.replace(' ', '_') if isinstance(x, (str, unicode)) else x)
df.columns = cols

df.to_csv("../data/data.csv")
```
